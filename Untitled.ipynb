```python
import pandas as pd
import os

# 定义文件路径，已根据你提供的路径进行修改
file_path = r'C:\Users\crj\overstimulation_dataset.csv'

# 检查文件是否存在，防止运行报错
if not os.path.exists(file_path):
    print(f"错误：文件 '{file_path}' 不存在。请检查文件路径是否正确。")
else:
    # 1. 加载数据集
    try:
        df = pd.read_csv(file_path)
        print("数据集加载成功！")

        # 2. 筛选出与论文相关的列
        # 这些列包含了睡眠指标、可能影响睡眠的环境/行为因素
        relevant_columns = [
            'Sleep_Hours',           # 睡眠时长，与你的“总睡眠时间”相关
            'Sleep_Quality',         # 睡眠质量，与你的“睡眠质量评分”相关
            'Screen_Time',           # 屏幕使用时间，与光照/蓝光危害间接相关
            'Tech_Usage_Hours',      # 技术使用时长，与光照/蓝光危害间接相关
            'Stress_Level',          # 压力水平，一个重要的睡眠影响因素
            'Anxiety_Score',         # 焦虑评分，可能影响睡眠
            'Depression_Score',      # 抑郁评分，可能影响睡眠
            'Environmental_Noise'    # 环境噪音，一个重要的环境因素
        ]

        # 检查所有筛选的列是否存在于数据集中
        existing_columns = [col for col in relevant_columns if col in df.columns]
        if len(existing_columns) < len(relevant_columns):
            missing_cols = list(set(relevant_columns) - set(existing_columns))
            print(f"警告：以下列在数据集中不存在，已跳过：{missing_cols}")
        
        # 如果筛选后没有剩下任何列，则退出
        if not existing_columns:
            print("\n错误：筛选后没有找到任何有效的列。请检查列名是否正确。")
        else:
            # 3. 创建一个新的DataFrame，只包含筛选后的列
            filtered_df = df[existing_columns]

            # 4. 显示新表格的基本信息，以供检查
            print("\n筛选后的数据表格（前5行）：")
            print(filtered_df.head())
            print("\n筛选后的表格信息：")
            filtered_df.info()

            # 5. 将筛选后的数据保存为新的CSV文件
            output_file_path = r'C:\Users\crj\filtered_overstimulation_sleep_data.csv'
            filtered_df.to_csv(output_file_path, index=False)
            print(f"\n筛选后的数据已成功保存到：{output_file_path}")

    except Exception as e:
        print(f"处理文件时发生错误：{e}")
```

```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt

# 设置中文显示
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 定义文件路径
file_path = r'C:\Users\crj\filtered_overstimulation_sleep_data.csv'

# 检查文件是否存在
if not os.path.exists(file_path):
    print(f"错误：文件 '{file_path}' 不存在。请检查文件路径是否正确。")
else:
    try:
        # 加载筛选后的数据集
        df = pd.read_csv(file_path)
        print("筛选后的数据集加载成功！")

        # 1. 计算相关系数矩阵
        print("\n--- 计算皮尔逊相关系数矩阵 ---")
        correlation_matrix = df.corr()
        print(correlation_matrix)

        # 2. 绘制相关性热力图
        print("\n--- 绘制相关性热力图 ---")
        plt.figure(figsize=(10, 8))
        sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f",
                    linewidths=.5, cbar_kws={'label': '相关系数'})
        plt.title('睡眠与环境因素相关性热力图', fontsize=15)
        plt.xticks(rotation=45, ha='right')
        plt.yticks(rotation=0)
        plt.tight_layout()
        
        # 保存热力图
        output_image_path = r'C:\Users\crj\correlation_heatmap.png'
        plt.savefig(output_image_path, dpi=300)
        print(f"相关性热力图已成功保存到：{output_image_path}")
        
        plt.show()

    except Exception as e:
        print(f"处理文件时发生错误：{e}")
```

```python
import pandas as pd
import os

# 定义所有数据文件所在的根目录
data_dir = r'E:\我的大学四年\大一下\数模\25年华数杯C题\其他数据'

# 定义问题4和问题1文件的完整路径
file_path_p4 = os.path.join(data_dir, '问题4.xls - Sheet1.csv')
file_path_p1 = os.path.join(data_dir, '问题1.xls - Sheet1.csv')

# 检查文件是否存在
if not os.path.exists(file_path_p4) or not os.path.exists(file_path_p1):
    print("错误：文件不存在。请检查路径和文件名是否正确。")
    print(f"检查路径: {file_path_p4}")
    print(f"检查路径: {file_path_p1}")
else:
    try:
        # --- 第一步：加载和转换核心数据（问题4） ---
        
        # 加载问题4数据，处理两行表头
        df_p4_raw = pd.read_csv(file_path_p4, header=[0, 1])
        print("问题4数据加载成功！")
        
        # 加载问题1数据，用于获取指标名称
        # 你的问题1文件可能包含了指标的详细列表
        df_p1_raw = pd.read_csv(file_path_p1)
        # 假设第一列是指标名称，去掉第一行（可能是表头）
        metrics = df_p1_raw.iloc[:, 0].tolist()
        
        # 重塑数据：将宽表转换为长表
        data_list = []
        for col_name in df_p4_raw.columns:
            subject_id = col_name[0]
            night_number = col_name[1]

            if 'Unnamed' in subject_id or 'Unnamed' in night_number:
                continue

            # 确保指标名称和数据长度匹配
            if len(df_p4_raw[col_name].values) != len(metrics):
                print(f"警告：被试ID {subject_id}，夜晚 {night_number} 的数据行数与指标列表不匹配。已跳过。")
                continue

            for i, value in enumerate(df_p4_raw[col_name].values):
                data_list.append({
                    '被试ID': subject_id,
                    '夜晚编号': night_number,
                    '指标名称': metrics[i],
                    '指标数值': value
                })
        
        df_p4_long = pd.DataFrame(data_list)
        
        # 数据清洗
        df_p4_long['指标数值'] = pd.to_numeric(df_p4_long['指标数值'], errors='coerce')
        df_p4_long = df_p4_long.dropna().reset_index(drop=True)
        
        print("\n--- 问题4数据转换成功，前5行数据如下：---")
        print(df_p4_long.head())

        # --- 第二步：关联光照环境信息 ---
        
        # 创建一个字典，用于将夜晚编号映射到光照环境
        # 根据C题题目，这里做一个合理的实验设计假设
        night_to_light = {
            'Night 1': '普通光照',
            'Night 2': '黑暗环境',
            'Night 3': '优化光照'
        }

        # 使用字典进行映射，创建新的“光照环境”列
        df_p4_long['光照环境'] = df_p4_long['夜晚编号'].map(night_to_light)
        
        # 打印整合后的数据前5行，查看新的'光照环境'列
        print("\n--- 整合光照环境后的数据前5行 ---")
        print(df_p4_long.head())
        
        # 将最终整合后的数据保存到新的CSV文件
        output_path_final = os.path.join(data_dir, 'final_integrated_data.csv')
        df_p4_long.to_csv(output_path_final, index=False)
        print(f"\n最终整合后的数据已成功保存到：{output_path_final}")

    except Exception as e:
        print(f"处理文件时发生错误：{e}")
```
